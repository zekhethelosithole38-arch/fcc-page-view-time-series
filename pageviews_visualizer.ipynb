```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/zekhe/Downloads/fcc-forum-pageviews.csv") #fix the date issue
```

```python
df = df[((df["value"] >= df["value"].quantile(0.025)) & (df["value"] <= df["value"].quantile(0.975)))]
```

```python
plt.plot(df["date"], df["value"], color = "red")
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel("Date")
plt.ylabel("Page Views")
```

```python
df["date"] = pd.to_datetime(df["date"])
df_bar = df.groupby([df["date"].dt.year, df["date"].dt.month])["value"].mean().unstack()
```

```python
fig = df_bar.plot(kind="bar").figure
plt.xlabel("Years")
plt.ylabel("Average Page Views")
plt.legend(title="Months", labels=["January","February","March","April","May","June","July","August","September","October","November","December"])
```

```python
fig = df_bar.plot(kind="bar").figure
plt.xlabel("Years")
plt.ylabel("Average Page Views")
plt.legend(title="Months", labels=[
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
])
```

```python
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]

sns.boxplot(x = "year", y = "value", data = df_box, palette="husl")
plt.title("Year-wise Box Plot (Trend)")
plt.xlabel("Year")
plt.ylabel("Page Views")
plt.show()

sns.boxplot(x = "month", y = "value", data = df_box, order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"], palette = "husl")
plt.title("Month-wise Box Plot (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Page Views")

```
